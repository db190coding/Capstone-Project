### Connect to the database

Let us first load the SQL extension and establish a connection with the database


```python
#I worked with my own Oracle SQL database instead of IBM db2, due to service problems, therefore I imported relevant classes 
import cx_Oracle
import json
import pandas as pd

class Json:
    def __init__(self, path):
        self.path = path
    
    # Function for reading a Json file
    def read(self,path):
        with open(path,"r") as j:
            config = json.load(j)
        return config

class Database():
    def __init__(self):
        config_path = "db_config.json" # path to the Json config file (needs to be in the same location as the script)
        config_file = Json(config_path) # Creation of a Json object
        conf_dict = config_file.read(config_file.path) # Reads Json to a dictionary
        self.user = conf_dict["db_username"]
        self.password = conf_dict["db_password"]
        self.host = conf_dict["db_ip"]
        self.port = conf_dict["db_port"]
        self.db = conf_dict["db_name"]
        self.conn = self.connect()

    # function for connecting to tha database
    def connect(self):
        dsn = cx_Oracle.makedsn(self.host, self.port, self.db)
        conn = cx_Oracle.connect(self.user, self.password, dsn)

        return conn

    def read(self, sql_command):
        result = pd.read_sql(sql_command,self.conn)
        return result
```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
db.read("SELECT DISTINCT LAUNCH_SITE FROM V_SPACEX")
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
db.read("SELECT * FROM V_SPACEX where LAUNCH_SITE LIKE 'CCA%' and rownum <= 5")
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
db.read("SELECT SUM(PAYLOAD_MASS__KG_) FROM V_SPACEX where CUSTOMER = 'NASA (CRS)'")
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
db.read("SELECT AVG(PAYLOAD_MASS__KG_) FROM V_SPACEX where BOOSTER_VERSION = 'F9 v1.1'")
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
db.read("SELECT MIN(DATE_) FROM V_SPACEX where LANDING_OUTCOME = 'Success (ground pad)'")
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
db.read("SELECT DISTINCT BOOSTER_VERSION FROM V_SPACEX WHERE LANDING_OUTCOME = 'Success (drone ship)' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000")
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
db.read("""SELECT MISSION_OUTCOME, COUNT(*) FROM (
SELECT SUBSTR(MISSION_OUTCOME,0,7) as MISSION_OUTCOME FROM V_SPACEX)
GROUP BY MISSION_OUTCOME""")
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
db.read("SELECT DISTINCT BOOSTER_VERSION FROM V_SPACEX WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM V_SPACEX)")
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
db.read("SELECT BOOSTER_VERSION, LAUNCH_SITE FROM V_SPACEX WHERE LANDING_OUTCOME = 'Failure (drone ship)' AND EXTRACT(YEAR FROM DATE_) = '2015'")
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
db.read("SELECT LANDING_OUTCOME, COUNT(*) FROM V_SPACEX WHERE DATE_ BETWEEN '04.06.2010' AND '20.03.2017' GROUP BY LANDING_OUTCOME ORDER BY COUNT(*) desc")
```
